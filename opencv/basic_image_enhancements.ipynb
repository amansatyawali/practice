```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
landscape_bgr = cv2.imread("landscape.jpg")
plt.imshow(landscape_bgr[ : , : , ::-1])
landscape_shape = landscape_bgr.shape
print(landscape_shape)
```

```python
#increas and decrease brightness

matrix = np.ones(landscape_shape, dtype = 'uint8')*50
bright_landscape_bgr = cv2.add(landscape_bgr, matrix)
dull_landscape_bgr = cv2.subtract(landscape_bgr, matrix)

plt.figure(figsize = [18, 5])
plt.subplot(131); plt.imshow(bright_landscape_bgr[ : , : , ::-1]) ; plt.title('Lighter')
plt.subplot(132); plt.imshow(landscape_bgr[ : , : , ::-1]) ; plt.title('Original')
plt.subplot(133); plt.imshow(dull_landscape_bgr[ : , : , ::-1]) ; plt.title('Darker')
```

```python
#increas and decrease Contrast


low_intensity_matrix = np.ones(landscape_shape, dtype = 'uint8')*0.8
high_intensity_matrix = np.ones(landscape_shape, dtype = 'uint8')*1.2

low_contrast_landscape_bgr = np.uint8(cv2.multiply(np.float64(landscape_bgr), low_intensity_matrix))
high_contrast_landscape_bgr = np.uint8(cv2.multiply(np.float64(landscape_bgr), high_intensity_matrix))

plt.figure(figsize = [18, 5])
plt.subplot(131); plt.imshow(low_contrast_landscape_bgr[ : , : , ::-1]) ; plt.title('Low contrast')
plt.subplot(132); plt.imshow(landscape_bgr[ : , : , ::-1]) ; plt.title('Original')
plt.subplot(133); plt.imshow(high_contrast_landscape_bgr[ : , : , ::-1]) ; plt.title('High contrast')
```

```python
#increas and decrease Contrast while handling the pixels that are too bright


low_intensity_matrix = np.ones(landscape_shape, dtype = 'uint8')*0.8
high_intensity_matrix = np.ones(landscape_shape, dtype = 'uint8')*1.2

low_contrast_landscape_bgr = np.uint8(cv2.multiply(np.float64(landscape_bgr), low_intensity_matrix))
high_contrast_landscape_bgr = np.uint8(np.clip(cv2.multiply(np.float64(landscape_bgr), high_intensity_matrix), 0, 255))

plt.figure(figsize = [18, 5])
plt.subplot(131); plt.imshow(low_contrast_landscape_bgr[ : , : , ::-1]) ; plt.title('Low contrast')
plt.subplot(132); plt.imshow(landscape_bgr[ : , : , ::-1]) ; plt.title('Original')
plt.subplot(133); plt.imshow(high_contrast_landscape_bgr[ : , : , ::-1]) ; plt.title('High contrast')
```

```python
retval, img_thresh = cv2.threshold(landscape_bgr, 100, 255, type = cv2.THRESH_BINARY)
print(retval)
plt.imshow(img_thresh[ : , : , ::-1])
```

```python
handwriting = cv2.imread("hand_written.jpg", 0)
plt.imshow(handwriting, cmap = 'gray')
```

```python
retval, handwriting_thresh_1 = cv2.threshold(handwriting, 100, 255, type = cv2.THRESH_BINARY)

retval, handwriting_thresh_2 = cv2.threshold(handwriting, 50, 255, type = cv2.THRESH_BINARY)

adaptive_thresh_handwriting = cv2.adaptiveThreshold(handwriting, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 51, 7)

plt.figure(figsize = [18, 15])
plt.subplot(221); plt.imshow(handwriting, cmap = 'gray') ; plt.title('Original')
plt.subplot(222); plt.imshow(handwriting_thresh_1, cmap = 'gray') ; plt.title('Threshold = 100')
plt.subplot(223); plt.imshow(handwriting_thresh_2, cmap = 'gray') ; plt.title('Threshold = 50')
plt.subplot(224); plt.imshow(adaptive_thresh_handwriting, cmap = 'gray') ; plt.title('Adaptive threshold')
```
