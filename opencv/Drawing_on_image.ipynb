```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
eifel_bgr = cv2.imread("eifel_tower.jpg", 1)
print(eifel_bgr.shape)
```

```python
plt.imshow(eifel_bgr[ : , : , ::-1])
```

```python
lined_image = eifel_bgr.copy()
cv2.line(lined_image, pt1 = (0, 0),pt2 = (100,200), color = (0, 255, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(lined_image[ : , : , ::-1])
```

```python
circle_on_image = eifel_bgr.copy()
cv2.circle(circle_on_image, center = (100,200), radius = 150, color = (255, 255, 0), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(circle_on_image[ : , : , ::-1])
```

```python
rectangle_on_image = eifel_bgr.copy()
cv2.rectangle(rectangle_on_image, pt1 = (50, 100), pt2 = (150, 300), color = (255, 0, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(rectangle_on_image[ : , : , ::-1])
```

```python
text_on_image = eifel_bgr.copy()
text_val = 'Eifel Tower, Paris'
font_scale = 2.3
font_face = cv2.FONT_HERSHEY_PLAIN
font_color = (255, 255, 0)
font_thickness = 2
cv2.putText(text_on_image, text_val,(100, 300), font_face, font_scale, font_color, font_thickness, lineType = cv2.LINE_AA)
plt.imshow(text_on_image[ : , : , ::-1])
```
