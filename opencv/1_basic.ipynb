```python
landscape_bgr = cv2.imread("landscape.jpg")
```

```python
landscape_b, landscape_g, landscape_r = cv2.split(landscape_bgr)

#Show channels
plt.figure(figsize = [20, 5])
plt.subplot(141);plt.imshow(landscape_r, cmap = 'gray');plt.title("Red channel");
plt.subplot(142);plt.imshow(landscape_g, cmap = 'gray');plt.title("Green channel");
plt.subplot(143);plt.imshow(landscape_b, cmap = 'gray');plt.title("Blue channel");

#Merge the individual channels into a BGR image

imMerged = cv2.merge((landscape_b, landscape_g, landscape_r))

#Show the merged output

plt.subplot(144);plt.imshow(imMerged[:, :, ::-1]);plt.title("Merged output");
```

```python
#Changing the color space of the image

#Changing BGR to RGB
landscape_rgb = cv2.cvtColor(landscape_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(landscape_rgb)
```

```python
#Changing BGR to HSV
landscape_hsv = cv2.cvtColor(landscape_bgr, cv2.COLOR_BGR2HSV)

landscape_h, landscape_s, landscape_v = cv2.split(landscape_hsv)

#Show channels
plt.figure(figsize = [20, 5])
plt.subplot(141);plt.imshow(landscape_h, cmap = 'gray');plt.title("Hue channel");
plt.subplot(142);plt.imshow(landscape_s, cmap = 'gray');plt.title("Saturation channel");
plt.subplot(143);plt.imshow(landscape_v, cmap = 'gray');plt.title("Value channel");

#Merge the individual channels into a BGR image

imMerged = cv2.merge((landscape_h, landscape_s, landscape_v))

#Show the merged output

plt.subplot(144);plt.imshow(cv2.cvtColor(imMerged, cv2.COLOR_HSV2BGR)[:, :, ::-1]);plt.title("Merged output");
```

```python
#Increasing hue by 10

landscape_h = landscape_h + 10


#Show channels
plt.figure(figsize = [20, 5])
plt.subplot(141);plt.imshow(landscape_h, cmap = 'gray');plt.title("Hue channel");
plt.subplot(142);plt.imshow(landscape_s, cmap = 'gray');plt.title("Saturation channel");
plt.subplot(143);plt.imshow(landscape_v, cmap = 'gray');plt.title("Value channel");

#Merge the individual channels into a BGR image

imMerged = cv2.merge((landscape_h, landscape_s, landscape_v))

#Show the merged output

plt.subplot(144);plt.imshow(cv2.cvtColor(imMerged, cv2.COLOR_HSV2BGR)[:, :, ::-1]);plt.title("Merged output");
```

```python
#Writing the image in a new file

landscape_increased_hue_bgr = cv2.cvtColor(imMerged, cv2.COLOR_HSV2BGR)
cv2.imwrite("landscape_increased_hue.jpg", landscape_increased_hue_bgr)
```
