```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bird_bgr = cv2.imread("index.png")
```

```python
plt.imshow(bird_bgr[ : , : , ::-1])
```

```python
enlarged_bird = cv2.resize(bird_bgr, None, fx = 2, fy = 3)
plt.imshow(enlarged_bird[ : , : , ::-1])
```

```python
width = 400
height = 500

dim = (width, height)
enlarged_bird = cv2.resize(bird_bgr, dsize = dim)
plt.imshow(enlarged_bird[ : , : , ::-1])
```
