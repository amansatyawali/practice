```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
fullDf = pd.read_csv('data/twcs/twcs.csv', nrows = 100000)
```

```python
df = fullDf[['text']]
df
```

```python
# df['text'] = df['text'].astype(str)
df['text_lower'] = df['text'].str.lower()

df
```

```python
stop_words = set(stopwords.words('english'))

def removeStopWords(text) :
    text_tokens = nltk.tokenize.word_tokenize(text)
    text = ' '.join([i for i in text_tokens if i not in stop_words])
    return text
```

```python
df['text_no_stopwords'] = df['text_lower'].apply(lambda text : removeStopWords(text))
```

```python
PUNCT = set(string.punctuation)

def removePunctuations(text) :
    for i in text :
        if i in PUNCT :
            text = text.replace(i, '')
    return text
```

```python
df['text_no_punct'] = df['text_no_stopwords'].apply(lambda text : removePunctuations(text))
```

```python
wordnet_map = {
    #'CC' : None,
    #'CD' : None,
    #'DT' : None,
    #'EX' : None,
    #'FW' : None,
    #'IN' : None,
    'JJ' : 'a',
    'JJR' : 'a',
    'JJS' : 'a',
    #'LS' : None,
    #'MD' : None,
    'NN' : 'n',
    'NNS' : 'n',
    'NNP' : 'n',
    'NNPS' : 'n',
    #'PDT' : None,
    #'POS' : None,
    #'PRP' : None,
    '#PRP$' : None,
    'RB' : 'r',
    'RBR' : 'r',
    'RBS' : 'r',
    #'RP' : None,
    #'TO' : None,
    #'UH' : None,
    'VB' : 'v',
    'VBG' : 'v',
    'VBD' : 'v',
    'VBN' : 'v',
    'VBP' : 'v',
    'VBZ' : 'v',
    #'WDT' : None,
    #'WP' : None,
    'WRB' : 'r' 
  }

lemma = WordNetLemmatizer()


def lemmatizeText(text) : 
    pos = nltk.pos_tag(text.split())
    text = ' '.join([lemma.lemmatize(i[0], wordnet_map.get(i[1], 'n')) for i in pos])
    return text
```

```python
df['text_lemma'] = df['text_no_punct'].apply(lambda x : lemmatizeText(x))
```
